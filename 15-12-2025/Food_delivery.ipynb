# Exercise 15

Which cuisine generates the highest order_amount overall?

```python
from pyspark.sql.functions import sum

total_order_amount_per_cuisine_df = df.groupBy('cuisine').agg(sum('order_amount').alias('total_order_amount'))
total_order_amount_per_cuisine_df.show()
```

```python
highest_order_cuisine_df = total_order_amount_per_cuisine_df.orderBy(desc('total_order_amount')).limit(1)
highest_order_cuisine_df.show()
```

# Exercise 16

Which city has the highest number of orders?

```python
from pyspark.sql.functions import count, desc

city_order_counts_df = df.groupBy('city').agg(count('order_id').alias('total_orders'))
highest_order_city_df = city_order_counts_df.orderBy(desc('total_orders')).limit(1)
highest_order_city_df.show()
```

# Exercise 17

Which payment mode is most frequently used?

```python
from pyspark.sql.functions import count, desc

payment_mode_counts_df = df.groupBy('payment_mode').agg(count('order_id').alias('usage_count'))
most_frequent_payment_mode_df = payment_mode_counts_df.orderBy(desc('usage_count')).limit(1)
most_frequent_payment_mode_df.show()
```

# Exercise 18

Why is Parquet the preferred format for analytics platforms like Databricks and BigQuery?



Parquet is a columnar storage format, and its preference in analytics platforms like Databricks and BigQuery stems from several key advantages:

Columnar Storage: Unlike row-oriented formats (like CSV or JSON), Parquet stores data column by column. This is highly efficient for analytical queries because queries often involve only a subset of columns. The engine can read only the necessary columns, significantly reducing I/O operations.
Compression: Parquet offers superior data compression compared to row-oriented formats. Because data in a single column is typically of the same type and often has similar values, it compresses much better. This saves storage space and further reduces I/O when reading data.
Encoding Schemes: It supports various encoding schemes (e.g., dictionary encoding, run-length encoding) that are optimized for columnar data, contributing to better compression and query performance.

Predicate Pushdown: Analytics engines can "push down" filters to the Parquet reader. This means that data can be filtered at the storage level, allowing the engine to skip reading entire blocks or pages of data that don't match the filter criteria. This drastically speeds up queries.

Schema Evolution: Parquet supports schema evolution, allowing users to add new columns, rename columns, and evolve the schema over time without breaking existing data or queries.
Optimized for Analytics Workloads: Its design is specifically tailored for OLAP (Online Analytical Processing) workloads, where complex queries operate on large datasets. It's not ideal for OLTP (Online Transaction Processing) where individual row lookups and modifications are frequent.

Integration with Big Data Ecosystem: Parquet is a cornerstone of the Apache Hadoop ecosystem and integrates seamlessly with Spark, Hive, Impala, and other big data tools, making it a natural fit for platforms built on these technologies.

Cost-Efficiency: Reduced storage and faster query execution directly translate to lower operational costs on cloud platforms, as you pay less for storage and compute time.
